#### (6 marks) Write a function take_biscuits(b_list, b_name, b_number) to allow multiple biscuits to be taken at once. Your function should:
1. Check if there are enough biscuits (b_number) of that type (b_name) in
the biscuit list (b_list)
    * Print a confirmation if there are enough, and delete the biscuits
    * Give an error message if not.
2. Return the list of selected biscuits, or an empty list if not enough.


````python
def take_biscuits(b_list, b_name, b_number):
    if b_name not in b_list or b_number >= b_list.count(b_name):
        print(f'Sorry, we do not have enough {b_name} biscuits.')
        return []

    print(f'You have selected {b_number} {b_name} biscuits.')
    selected_biscuits = []
    for _ in range(b_number):
        if b_name in b_list:
            b_list.remove(b_name)
            selected_biscuits.append(b_name)
    return selected_biscuits

take_biscuits(biscuits, 'Kingston', 3)
````

#### i) (5 marks) Write a Dog class – held in dogs.py, with instance variables for breed and four criteria: size, friendly, clever, strong (ratings (1-5)). Include methods for intialising the object and converting it to a string. See UML class diagram, below.



```

-------------------------------------------------
```

````python
##dogs.py
class Dog:
    def __init__(self, breed, size, friendly, clever, strong):
        self.breed = breed
        self.size = self._validate_rating(size, "Size")
        self.friendly = self._validate_rating(friendly, "Friendly")
        self.clever = self._validate_rating(clever, "Clever")
        self.strong = self._validate_rating(strong, "Strong")

    def _validate_rating(self, value, attribute_name):
        if isinstance(value, int) and 1 <= value <= 5:
            return value
        else:
            raise ValueError(f"{attribute_name} rating must be an integer between 1 and 5.")

    def __str__(self):
        return (f"Dog\n\nBreed:\t\t{self.breed}\n"
                f"\nCriteria\tRating\n"
                f"Size:\t\t{self.size}/5\n"
                f"Friendly:\t{self.friendly}/5\n"
                f"Clever:\t\t{self.clever}/5\n"
                f"Strong:\t\t{self.strong}/5")
````

ratings

````python
##dogRating.py
# from dogs import Dog

def main():
    dogs = []
    while True:
        try:
            breed = input("Enter dog breed (or type 'done' to finish): ")
            if breed.lower() == 'done':
                break

            size = int(input("Enter size rating (1-5): "))
            friendly = int(input("Enter friendly rating (1-5): "))
            clever = int(input("Enter clever rating (1-5): "))
            strong = int(input("Enter strong rating (1-5): "))

            dog = Dog(breed, size, friendly, clever, strong)
            dogs.append(dog)

        except ValueError as err:
            print(f"Invalid input: {err}")
        except Exception as err:
            print(f"An unexpected error occurred: {err}")

    print("\n--- Dog Ratings ---")
    for dog in dogs:
        print(dog)

if __name__ == "__main__":
    main()
````

#### e) (2 marks). Using a diagram, show the difference between von Neumann neighbourhoods and Moore neighbourhoods and indicate which was used in the heat diffusion example

````python
import matplotlib.pyplot as plt

# Von Neumann Neighborhood
# A cell's neighbors are its immediate up, down, left and right neighbors.

# Create a sample grid to visualize the neighborhood
grid = [[0, 0, 0],
        [0, 1, 0],
        [0, 0, 0]]

fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Plot the grid for von Neumann neighborhood
axes[0].imshow(grid, cmap='viridis')
axes[0].set_title("Von Neumann Neighborhood")
axes[0].set_xticks([])  # Hide x-ticks
axes[0].set_yticks([])  # Hide y-ticks
for i in range(3):
    for j in range(3):
      if grid[i][j]==1:
        axes[0].text(j,i, '1', ha="center", va="center", color="w")


# Moore Neighborhood
# A cell's neighbors are all cells surrounding it, including diagonals.
grid_moore = [[1,1,1],
        [1,1,1],
        [1,1,1]]
grid_moore[1][1] = 0
# Plot the grid for Moore neighborhood
axes[1].imshow(grid_moore, cmap='viridis')
axes[1].set_title("Moore Neighborhood")
axes[1].set_xticks([])  # Hide x-ticks
axes[1].set_yticks([])  # Hide y-ticks
for i in range(3):
    for j in range(3):
        if grid_moore[i][j] == 0:
          axes[1].text(j,i, '0', ha="center", va="center", color="w")

plt.tight_layout()
plt.show()

print("In the heat diffusion example from the code, the Von Neumann neighborhood is used")
````
